# Import General Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data

```python
df = pd.read_csv('Downloads/data_lokasi_busway_jakarta_latlong.csv', index_col=None, header=0)
df
```

```python
df2 = pd.read_csv('Downloads/data_jumlah_penumpang_2020.csv', index_col=None, header=0)
df2
```

```python
df_merge = pd.merge(df, df2, on="koridor_no")
df_merge
```

# Initialize KeplerGl

```python
from keplergl import KeplerGl 
map_1 = KeplerGl(height=800, width=800)
map_1
```

```python
df_test = pd.DataFrame()
df_test = df_merge
# df_merge[df_merge['month']=='2020-01-01']
df_test
```

```python
import geopandas as gpd

# Create a gepdataframe
gdf = gpd.GeoDataFrame(df_merge, geometry=gpd.points_from_xy(df_merge.long, df_merge.lat))
# Add data to Kepler
map_1.add_data(data=gdf, name="jumlah_penumpang")

# map_1.add_data(data=df2, name="data_1")
```

```python
map_1
```

```python
config = map_1.config
config
```

```python
# Save current map directly
map_1.save_to_html(file_name="my_keplergl_map.html")

# Save map using other data or config
map_1.save_to_html(data={'data_1': df_merge}, config=config, file_name="my_keplergl_map.html")
```
